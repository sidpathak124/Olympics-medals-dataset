# Do GDP, Population and Politics matter ?

### Our hypothesis are :-

Higher GDP should lead to More medals

Higher Population should lead to More medals

More stable government should lead to higher medals

Also, number of games that a country participate in has an influence on the number of medal it wins (No Brainer!)

let us test this

###  Let us first prepare our data

```python
# Total medals won by each country
medals_per_country = pd.crosstab(index = olympics.Country, columns = olympics.Medal, margins = True, margins_name = 'Total')
medals_per_country
```

```python
medals_per_country.drop('Total', axis = 0, inplace = True)
medals_per_country
```

```python
# rearranging the df
medals_per_country = medals_per_country[['Total', 'Gold', 'Silver', 'Bronze']].sort_values(by = 
                    ['Total', 'Gold', 'Silver', 'Bronze'], ascending = False)
```

```python
medals_per_country
```

```python
dicty
```

```python
# adding 'Population' and 'GDP' cols to our 'medals_per_country' df

medals_per_country = medals_per_country.merge(dicty, how = 'left', left_index = True, right_on = 'Country').set_index(
    'Country').drop(columns = 'Code')
```

```python
medals_per_country
```

## Statistical analysis and hypothesis testing

```python
medals_per_country
```

```python
# we need only total columns here
medals_per_country.drop(columns = ['Gold', 'Silver', 'Bronze'], inplace = True)
```

```python
medals_per_country.corr(method = 'pearson')
# shows a linear correlation
```

```python
medals_per_country.corr(method = 'spearman')
# shows even greater correlation between total medals and Population, GDP, Total_games played
```

```python
# spearman is nothing but pearson correlation on ranks (rather than absolute values)
medals_per_country.rank().corr(method = 'pearson')
```

### Hypothesis 1(Null): There is no relation between total medals and population 

```python
# calculating the p-value of the spearman test
import scipy.stats as stats

stats.spearmanr(medals_per_country.Total, medals_per_country.Population, nan_policy = 'omit')

# output is both 'correlation coefficient' and 'pvalue'
```

```python
r, pvalue = stats.spearmanr(medals_per_country.Total, medals_per_country.Population, nan_policy = 'omit')
```

```python
r
```

```python
pvalue
```

#### Thus, we reject the null hypothesis, there is a relation between total medals and population 

### Hypothesis 2(Null): There is no relation between total medals and GDP

```python
r, pvalue = stats.spearmanr(medals_per_country.Total, medals_per_country.GDP, nan_policy = 'omit')
```

```python
r
```

```python
pvalue
```

#### Thus, we reject the null hypothesis here as well


### Hypothesis 3(Null): There is no relation between total medals and Participations 

```python
r, pvalue = stats.spearmanr(medals_per_country.Total, medals_per_country.Total_games, nan_policy = 'omit')
```

```python
r
```

```python
pvalue
```

#### Thus, we reject the null hypothesis here as well 

## Aggregating and Ranking 

```python
sum_vs_win = pd.crosstab(olympics.Country, olympics.Edition)
sum_vs_win
```

```python
men_vs_women = pd.crosstab(olympics.Country, olympics.Gender)
men_vs_women
```

```python
medals_by_cats = pd.concat([sum_vs_win, men_vs_women], axis = 1)
medals_by_cats
```

```python
medals_by_cats['Total'] = medals_by_cats.summer + medals_by_cats.winter
medals_by_cats
```

```python
medals_by_cats.sort_values(by = 'Total', ascending = False, inplace = True)
medals_by_cats
```

```python
# Since the no. of events in "summer and winter" as well as for 'males and females' are very unequal
# it is better to use rank
ranks = medals_by_cats.rank(ascending = False, method = 'average')
ranks
```

```python
top_50 = ranks.head(50)
```

```python
# visualizing with a heatmap
plt.figure(figsize = (50, 10))
sns.heatmap(top_50.T, cmap = 'RdYlGn_r', annot = True, fmt = '2.0f')
plt.show()
```

# Summer Games vs Winter Games - does geographical location matter ? 

```python
# Here we can find the countries that are more summer oriented through the contrast
plt.figure(figsize = (50, 10))
sns.heatmap(top_50.sort_values(by = 'summer').T, cmap = 'RdYlGn_r', annot = True, fmt = '2.0f')
plt.show()
```

```python
# similarly for winter
plt.figure(figsize = (50, 10))
sns.heatmap(top_50.sort_values(by = 'winter').T, cmap = 'RdYlGn_r', annot = True, fmt = '2.0f')
plt.show()
```

```python
# To visualize the contrast in a better way

rank_diff = top_50.summer.sub(top_50.winter).sort_values().to_frame()
rank_diff
```

```python
plt.figure(figsize = (50, 10))
sns.heatmap(rank_diff.T, cmap = 'RdBu', annot = True, fmt = '2.0f', center = 0)
plt.show()

# countries on the right are pro-winter and the ones at left are pro-summer
```
